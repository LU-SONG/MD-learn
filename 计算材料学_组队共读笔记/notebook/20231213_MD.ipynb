### 3.3 周期性边界条件


#### 概念

**为什么要用周期性边界条件？**→ 真实体系总原子数10^24^，表面原子占比10^-8^，但是在盒子中1000个、10^6^原子体系表面（10%，1%）影响不可忽略 → 构建周期性边界条件，**消除表面效应** → 模拟体相性质
