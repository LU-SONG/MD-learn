## 宋璐-第7天-周期性边界与近邻列表
### 课后巩固
**Q1**：
**试述自由边界条件和周期性边界条件的差别。**

**A1**：
自由边界条件：在自由边界条件下，系统的边界不受任何外部干扰，物理量（如粒子、温度、压力等）在边界上可以自由变化。
适用于模拟与外部环境相互作用并且没有明确定义周期性结构的系统。


周期性边界条件：超晶胞单元在空间中像晶格一样平移填充，只需要有限个原子就可以填充好整个空间，当物体穿过一个边界时，它会在相邻边界上重新出现。可以有效消除表面效应。
在模拟晶体、周期性结构或大规模系统时非常有用。



**Q2**：
**什么是最小镜像约定？并画简图说明在周期性边界条件下不满足该约定会导致什么问题。引入邻域列表对MD计算有什么优势？**

**A2**：
最小镜像约定是指，超晶胞的的最小边长必须大于两倍的原子间相互作用的截断距离。在这种情况下，我们通过一次计算就可以计算出原子之间的距离。任意一个原子只与最近的原子发生相互作用，大大简化能量和力的计算。
![最小镜像约定](https://github.com/LU-SONG/MD-learn/blob/main/%E8%AE%A1%E7%AE%97%E6%9D%90%E6%96%99%E5%AD%A6_%E7%BB%84%E9%98%9F%E5%85%B1%E8%AF%BB%E7%AC%94%E8%AE%B0/notebook/images/%E6%9C%80%E5%B0%8F%E9%95%9C%E5%83%8F.jpg?raw=true)

在计算相互作用时，通常会对势场进行有限距离的截断处理。我们需要一个合理的原子截断半径，在这范围内的原子之间有正常的相互作用，截断半径外面的原子相互作用为0，不计算能量和力。可以极大降低计算成本。
