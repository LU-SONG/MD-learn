#### 为什么用DPMD计算？
DFT结果准确，但是计算开销太大，例如 cp2k，vasp
MD计算效率较高，但是结果不可信，使用经验力场例如LJ, EAM, MEA等。

优点：使用第一性原理计算精度 + 接近经验力场的计算效率
使用机器学习拟合高维势能面

目标：数值近似第一性原理定义的能量函数，不需要求解高维函数函数
     只需要用神经网络学习 DFT计算结果
存在问题：网络固定，拓展性较低。

##### DPMD的适用条件
体系能量是由所有原子能量贡献的和
但是MD计算的体系能量是由截断半径内的所有原子确定的，忽略了其他区域的原子能量，忽略了长程相互作用。
DPLR是DPMD计算的一种方法，可以计算长程相互作用
